```python
from sklearn.pipeline import Pipeline
from sklearn.base import TransformerMixin
class CustomCategoryImputer(TransformerMixin):
    def __init__(self, cols=None):
        self.cols = cols
    def transform(self, df):
        train = df.copy()
        for col in self.cols:
            train[col].fillna(train[col].value_counts().index[0], inplace=True)
            return train
    def fit(self, *_):
        return self
```

```python
y_test=test['Promoted_or_Not']
X_test=test[['Targets_met','Previous_Award']]
y_train=train['Promoted_or_Not']
X_train=train[['Targets_met','Previous_Award']]
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
para_grid={'C':[0.001,0.01,0.1,1,10],'gamma':[0.001,0.01,0.1,1]}
grid=GridSearchCV(SVC(kernel='rbf'),para_grid,verbose=2)
grid.fit(X_train,y_train)
grid.svm=grid.predict(X_test)
```

```python
rate=np.asarray(grid.svm)
np.savetxt('tue.csv',rate,delimiter=",")
```

```python
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(y_test,grid.svm)
```
